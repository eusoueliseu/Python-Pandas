#### Vamos ver também o método groupby

```python
#Vamos importar o pandas e carregar o primeiro csv

import pandas as pd
bond = pd.read_csv('jamesbond.csv', sep = ';', index_col = 'Film')
bond
```

```python
#Ao alocar a lista de filmes na variável Bond escolhemos a coluna "Film" como índice da tabela.
#Mas e se eu quiser trocar a coluna de índice?

bond.set_index(keys = 'Actor')
```

```python
#Mas como retornamos o Data Frame para índices numéricos?

bond = bond.reset_index()
bond
```

```python
#Já sabemos selecionar algumas colunas
#Também sabemos selecionar linhas com base em critérios
#Mas como escolher linhas manualmente?

#Vamos colocar o nome do filme como índice novamente e ordenar a planilha com base no índice
bond = pd.read_csv('jamesbond.csv', sep = ';', index_col = 'Film')
bond = bond.sort_index()
bond.head(8)
```

```python
#Vamos usar o método loc e inserir o nome do filme que buscamos

bond.loc['Thunderball']

#A busca é feita com base no índice da tabela

#O output de índices duplicados é um dataframe com todos os itens correspondentes
```

```python
#Caso meu desejo seja selecionar um range de linhas (da mesma forma que fazíamos com listas e tuplas), basta adicionar o
#índice inicial e final da seleção

bond.loc['From Russia with Love' : 'Octopussy']
```

```python
#Além disso, de forma similar as listas, podemos selecionar linhas esparsas

bond.loc[['Live and Let Die', 'The Spy Who Loved Me', 'GoldenEye']]
```

```python
#Agora vamos fazer seleções com base no método iloc
#Este é utilizado para seleção por índices dos dados

#Vamos olhar a primeira linha do dataframe

bond.iloc[0]
```

```python
#Caso eu queira olhar mais de uma linha, basta passar uma lista dos índices que desejo acessar

bond.iloc[[1, 6, 12]]

#O output é um dataframe
```

```python
#Mas como passar argumentos a mais para os métodos loc e iloc?

#Com o método loc

#Vamos supor que eu queira ver apenas o ator e ano de um filme específico

bond.loc['Goldfinger', ['Actor','Director']]
```

```python
#Além de inserir uma lista com as colunas desejadas, podemos inserir múltiplos índices

bond.loc[['GoldenEye', 'Moonraker'], ['Actor', 'Director']]
```

```python
#Como fazer isso com o iloc?
#Usaremos os índices

bond.iloc[3, [1, 2]]

#Selecionamos um filme específico, bem como as colunas de ator e diretor
```

```python
#Selecionando valores desejados via índice

bond.iloc[[14, 21], [2, 4]]
```

```python
#Importante notar que tanto no LOC quanto no ILOC, o primeiro argumento refere-se a linha e o segundo a coluna
```

### Como inserir valores em células?

```python
#primeiro vamos olhar o valor atual do ator de "Licence to kill"

bond.loc['Licence to Kill', 'Actor']

#O output é Timothy Dalton, porém vamos mudar para Wagner Moura
```

```python
#Vamos subscrever o nome

bond.loc['Licence to Kill', 'Actor'] = 'Wagner Moura'

#Com o nome alterado, podemos acessar a série do filme e verificar o nome atualizado

bond.loc['Licence to Kill']
```

```python
#Agora vamos alterar mais de uma célula

bond.loc['Dr. No', ['Year', 'Actor', 'Director']] = ['2001', 'Seu Jorge', 'José Padilha']

#Com os valores alterados, vamos acessar a série

bond.loc['Dr. No']
```

## Groupby

#### Vamos olhar como funciona o Group by e como aproveitar essa função para facilitar nossas análises

```python
import pandas as pd
empresas = pd.read_csv('fortune1000.csv', index_col = 'Rank')
```

```python
#A tabela possui diversos valores duplicados na coluna de setor, então iremos agrupar as empresas
#a partir desta coluna

setores = empresas.groupby('Sector')
```

```python
#Vamos descobrir quantos grupos foram criados
len(setores)
```

```python
#Vamos olhar quantas empresas estão em cada setor
setores.size()
```

```python
#Caso precisemos olhar um item de cada setor, basta:

setores.first()

#setores.last faria a mesma coisa, porém com os últimos itens da lista
#Este comando é um ótimo overview da tabela
```

```python
#Como filtrar os valores do grupo?
#Vamos procurar todas empresas do setor de Transportes

setores.get_group('Transportation').head(8)

#O resultado será um dataframe com base na filtragem
```

```python
#Funções de agragação são possíveis com os grupos

setores.sum() #Este comando soma todos os valores dos grupos aninhados
setores.mean() #Este comando mostra a média dos valores para todos as empresas de cada setor
```

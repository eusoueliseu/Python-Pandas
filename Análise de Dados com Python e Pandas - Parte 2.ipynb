### Adicionando colunas ao dataframe

### Como excluir linhas, colunas e Missing Values?

##### Para isso vamos carregar outro dataset

```python
notas = pd.read_csv('notas.csv', sep = ';')

display(notas)

#Este dataset possui uma séries de missing values
```

```python
#Para lidar com esse problema vamos usar o drop NA

notas.dropna()

#Este comando excluiu as linhas que possuem qualquer nota com NA e manteve as outras
```

```python
#Mas se quisermos exluir apenas linhas com todos os valores "NA" e manter as outras linhas?

notas.dropna(how = 'all')

#Neste caso, nenhuma linha foi excluída, pois não há linhas com todos os valores faltantes
```

```python
#Caso precisemos remover as colunas com missing values, ao passo que mantemos as outras:

notas.dropna(axis = 1, how = 'any')

#Apenas a coluna com notas de inglês foi mantida, por ser a única que não possuía missing values
```

```python
#Esses dois comandos podem ser úteis em alguns casos, mas na maioria dos casos nosso interesse será excluir apenas a linha com missing value em uma coluna específica

#No comando dropna vamos inserir as colunas que queremos verificar se há missing values

notas.dropna(subset = ['Matematica'])

#Apenas a linha da Bruna foi removida, por ser a única com missing value na nota de Matemática
```

```python
#Vamos excluir as linhas com missing values nas colunas Matemática e Biologia

notas.dropna(subset = ['Matematica', 'Biologia'])

#Os NAs de História foram mantidos
```

### Ordenando a planilha

#### Vamos criar uma coluna com a somatória das notas e usá-la como determinante de ordem

```python
#Criando a coluna Total

notas['Total'] = notas['Historia'] + notas['Matematica'] + notas['Biologia'] + notas['Ingles']

#Vamos ver como ficou

display(notas)
```

```python
#Para linhas com NA, o Python colocou NA no valor da coluna.

#Vamos colocar 0 nos campos com NA para eliminar esse problema

notas = notas.fillna(0)

#Agora vamos recalcular a coluna de totais

notas['Total'] = notas['Historia'] + notas['Matematica'] + notas['Biologia'] + notas['Ingles']

display(notas)
```

```python
#Para ordenar a tabela a partir do total de notas, basta:

notas.sort_values('Total') #Foi ordenada da menor para maior

notas.sort_values('Total', ascending = False) #Ordenamos da maior para menor
```

```python
#A ordenação pode ser alfabética

notas.sort_values('Nome')
```

```python
#Vamos ordenar por uma matéria qualquer

notas.sort_values('Matematica')

#A nota 0 da Bruna era um NA, mas foi substituído por zero há alguns passos atrás.
```

```python
#Ao ordenar a tabela pelo total, alguns valores ficaram duplicados.

#Por isso vamos ver como ordenar a tabela com base em duas colunas ou mais

#A título de exemplo, vamos adotar a nota de Matemática como fator de desempate

notas.sort_values(['Total', 'Matematica'], ascending = False)
```

```python
#Ordenar a tabela por valores é bastante interessante para a tabela de ações

acoes.sort_values('Fechamento (R$)', ascending = False)
```

```python
#Talvez a variação percentual do ativo seja o fator mais interessante de ser observado

#Primeiro vamos criar uma coluna com o valor percentual de variação

acoes['Variação (%)'] = ((acoes['Fechamento (R$)'] / acoes['Abertura (R$)'] - 1) * 100)
```

```python
#Agora vamos ordenar a planilha com base na variação de preços desses ativos

acoes.sort_values('Variação (%)')

#Podemos ver os papéis ordenados de maiores perdas até melhores desempenhos
```

```python
#Outra forma de ordenar tanto as ações quanto as notas com base num ranking é usar o método rank()

notas['Ranking'] = notas['Total'].rank(ascending = False).astype('int')

notas.sort_values(['Ranking', 'Matematica'], ascending = [True, False])
```
